In this Jupyter notebook we show how intermediate probabilities and intermediate quantum states can be visualised using 
functions *show_all_probabilities()* and *show_all_intermediate_states()* of class *QuantumUtil*.
Both functions have an instance of Circuit as input parameter.

As we have seen before, the quantum state of a circuit of $N$ qubits can be represented as

$|\psi\rangle = \sum_{i=0}^{2^N-1} \alpha_i |i\rangle$,

where $|i\rangle$ represents the $i$-th basis state and $\alpha_i$ the probability amplitude corresponding to that basis state.

The coefficients $\alpha_i$ are complex numbers and satisfy the normalization condition

$\sum_{i=0}^{2^N-1} |\alpha_i|^2 = 1$.

Upon calling function *show_all_probabilities()* the values of the $|\alpha_i|^2$ are shown using colors with dark blue indicating $0$
and dark red indicating $1$.

When function *show_all_intermediate_states()* is called, the values of the complex numbers $\alpha_i$ are shown. In this case the
*magnitude* of $\alpha_i$ is indicated by color coding and the *phase* by the direction of the arrow.

Both functions have input parameters *show_description* and *show_colorbar* (not required). 
The flag *show_description* is used to show descriptions such as 'Hadamard on qubit 0' (default value is True) and the
flag *show_colorbar* is used to show the colorbar to the right of the figure (default value is True).

Import the QuantumSim code.

```python
from quantumsim import *
```

## Some examples
Below, a circuit of $4$ qubits is created. The Hadamard gate is applied to qubit $0$ followed by $3$ SWAP operations.
After executing and measuring the circuit $1000$ times, a histogram of resulting classical states is shown.

In addition, intermediate probabilities and quantum states are shown using functions *show_all_probabilities()* and *show_all_intermediate_states()*
with flags *show_description* and *show_colorbar* set to False as well as default settings.

```python
circuit = Circuit(4)
circuit.hadamard(0)
circuit.swap(0,1)
circuit.swap(1,2)
circuit.swap(2,3)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_probabilities(circuit, show_description=False, show_colorbar=False)
QuantumUtil.show_all_intermediate_states(circuit, show_description=False, show_colorbar=False)
QuantumUtil.show_all_probabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```

Below, several circuits are created. All of them are executed and measured. In all cases a histogram of classical states is shown as well as
intermediate probabilities and intermediate quantum states. Here, default settings for the flags *show_description* and *show_colorbar* are applied.

```python
circuit = Circuit(4)
circuit.pauli_x(0)
circuit.pauli_y(0)
circuit.hadamard(0)
circuit.cnot(0,1)
circuit.cnot(1,2)
circuit.cnot(2,3)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_probabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```

```python
circuit = Circuit(4)
circuit.pauli_x(0)
circuit.cnot(0,1)
circuit.controlled_pauli_y(0,2)
circuit.controlled_pauli_z(0,3)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_probabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```

```python
circuit = Circuit(4)
circuit.hadamard(0)
circuit.controlled_hadamard(0,1)
circuit.controlled_hadamard(1,2)
circuit.controlled_hadamard(2,3)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_probabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```

```python
circuit = Circuit(4)
for q in range(4):
    circuit.pauli_x(q)
circuit.hadamard(0)
circuit.controlled_hadamard(0,1)
circuit.controlled_hadamard(1,2)
circuit.controlled_hadamard(2,3)
result = QuantumUtil.run_circuit(circuit,2000)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_probabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```

```python
circuit = Circuit(4)
theta = np.pi/8
for q in range(4):
    circuit.pauli_x(q)
    circuit.phase(q*theta, q)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_probabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```

```python
circuit = Circuit(1)
circuit.pauli_x(0)
theta = np.pi/8
for q in range(16):
    circuit.phase(theta, 0)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_probabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```

```python
circuit = Circuit(5)
circuit.hadamard(1)
circuit.pauli_x(3)
circuit.fredkin(1,3,4)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_probabilities(circuit, show_description=False)
QuantumUtil.show_all_intermediate_states(circuit, show_description=False)
```

```python
circuit = Circuit(5)
circuit.hadamard(1)
circuit.hadamard(3)
circuit.toffoli(1,3,4)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_probabilities(circuit, show_description=False)
QuantumUtil.show_all_intermediate_states(circuit, show_description=False)
```
